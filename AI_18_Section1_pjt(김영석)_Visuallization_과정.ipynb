# 1. 지역별로 선호하는 장르가 다른가
시각화를 통해 확인한 결과 지역별로 다르다는 결과를 도출해낼 수 있었다.
우선 지역별을 나누는 기준은 장르를 기준으로 NA_Sales, EU_Sales, JP_Sales, Other_Sales를 각각 집계하였다. 장르의 기준이 많아 판매량 순위 상위 10개만 추려 확인하였다.

그래프 상으로 일본지역을 제외한 다른 지역에서는 액션 장르를 제일 선호하는 것을 확인하였다.

그럼 가설 검정을 통해 지역마다 선호하는 장르가 다른지 추가 확인이 필요하다.

* 귀무가설 : 지역마다 선호하는 장르가 같다.

* 대립가설 : 지역마다 선호하는 장르가 다르다.





```python
lc_Genre_NA = df.groupby(['Genre'])['NA_Sales'].sum()
lc_Genre_EU = df.groupby(['Genre'])['EU_Sales'].sum()
lc_Genre_JP = df.groupby(['Genre'])['JP_Sales'].sum()
lc_Genre_Other = df.groupby(['Genre'])['Other_Sales'].sum()
lc_Genre = pd.DataFrame()
lc_Genre = lc_Genre.append(lc_Genre_NA)
lc_Genre = lc_Genre.append(lc_Genre_EU)
lc_Genre = lc_Genre.append(lc_Genre_JP)
lc_Genre = lc_Genre.append(lc_Genre_Other)
lc_Genre = lc_Genre.T

from scipy.stats import chi2_contingency

chi2_val, p, dof, expected= chi2_contingency(lc_Genre, correction=False)
if(p<0.05) : 
    print('p value:', p,"\n"+'귀무가설을 기각한다. 지역마다 선호하는 장르가 다르다.')
else :
    print('p value:', p,"\n"+'귀무가설을 기각하지 못한다. 지역마다 선호하는 장르가 같다.')
```

카이제곱 검정을 통해 지역마다 선호하는 장르가 다르다는 점을 입증했다.

장르 상관없이 가장 많이 팔린 게임과 DS 기종에서 가장 많이 팔린 장르를 분석해 결합해보면 새로운 인사이트가 나오지 않을까

```python
# 장르 상관없이 가장 많이 팔린 게임
display(df.sort_values(by='Sales_sum', ascending=False))
# 장르가 스포츠, 기종은 wii , 명칭은 Wii Sports
# 해당 게임의 이력은 웨어러블 기기를 착용하여 실제 운동 동작을 인식하게 하여 비디오 게임을 즐기는 형식
# 연도별 판매 그래프를 보더라도 스포츠 장르는 순위권에 있음

# 상위 50개 스포츠 게임이 어느 지역에서 가장 잘 팔렸는지 확인
df_Sports = df[(df['Genre'] == 'Sports') & (df['Platform'] == 'PS4')]
df_top50_Sports = df_Sports.nlargest(50, 'Sales_sum')
df_sales = df_top50_Sports[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']]
# 지역별 총 판매량 계산
df_sales_sum = df_sales.sum()

# 파이차트 그리기
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
plt.pie(df_sales_sum, labels=df_sales.columns,colors=colors, autopct='%1.1f%%')
plt.title('Sports 장르 지역별 판매 비율')
plt.show()



```
